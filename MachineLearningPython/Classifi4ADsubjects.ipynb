```python
## EEG data
X = pd.read_csv('X307.csv', header=None)
y = pd.read_csv('Y307.csv', header=None)
y = np.ravel(y)
```

```python
print(X.shape)
```

```python

print(X.shape)
print(y.shape)
# Đếm số mẫu trong từng lớp
class_counts = pd.Series(y).value_counts()

# Hiển thị số mẫu trong từng lớp
print("Class Counts:")
print(class_counts)

# Tạo biến class_weight dựa trên số lượng mẫu trong từng lớp
class_weight = {label: class_counts.max() / count for label, count in class_counts.items()}

print("\nClass Weights:")
print(class_weight)
```

```python
num_folds = 10

# Định nghĩa các hyperparameters cho Logistic Regression
logistic_regression_param_grid = {
    'penalty': ['l1', 'l2'],
    'C': [0.001, 0.01, 0.1, 1, 10, 100]
}

# Mô hình Logistic Regression
logistic_regression_model = LogisticRegression(class_weight = class_weight)

# Tạo GridSearchCV object cho Logistic Regression
logistic_regression_grid_search = GridSearchCV(logistic_regression_model, logistic_regression_param_grid, cv=num_folds)

# Thực hiện tìm kiếm trên dữ liệu cho Logistic Regression
logistic_regression_grid_search.fit(X, y)

# In ra kết quả tốt nhất cho Logistic Regression
print("Best parameters for Logistic Regression: ", logistic_regression_grid_search.best_params_)
print("Best mean accuracy for Logistic Regression: {:.5f}".format(logistic_regression_grid_search.best_score_))

# Lấy mô hình tốt nhất từ GridSearchCV cho Logistic Regression
best_logistic_regression_model = logistic_regression_grid_search.best_estimator_

```

```python
# Định nghĩa các hyperparameters cho Random Forest
random_forest_param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Mô hình Random Forest
random_forest_model = RandomForestClassifier(class_weight = class_weight)

# Tạo GridSearchCV object cho Random Forest
random_forest_grid_search = GridSearchCV(random_forest_model, random_forest_param_grid, cv=num_folds)

# Thực hiện tìm kiếm trên dữ liệu cho Random Forest
random_forest_grid_search.fit(X, y)

# In ra kết quả tốt nhất cho Random Forest
print("\nBest parameters for Random Forest: ", random_forest_grid_search.best_params_)
print("Best mean accuracy for Random Forest: {:.5f}".format(random_forest_grid_search.best_score_))

# Lấy mô hình tốt nhất từ GridSearchCV cho Random Forest
best_random_forest_model = random_forest_grid_search.best_estimator_
```
